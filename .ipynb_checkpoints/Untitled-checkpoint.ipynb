```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import json

from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import sagemaker
```

```python
df = pd.read_csv('dataset.csv')
df.describe
```

```python
# df = shuffle(df, random_state=2)
```

```python
cols = [i for i in df.iloc[:,1:].columns]
cols
```

```python
tmp = pd.melt(df.reset_index() ,id_vars = ['index'], value_vars = cols )
tmp['add1'] = 1
tmp
```

```python

diseases = pd.pivot_table(tmp, 
                          values = 'add1',
                          index = 'index',
                          columns = 'value')

diseases.insert(0,'label',df['Disease'])
diseases = diseases.fillna(0)

diseases.head()
```

```python
diseaseLabels = diseases['label']
diseases = diseases.drop(columns = ['label'])

diseases.head()
```

```python
diseasesNames = diseaseLabels.unique();
diseasesNames = pd.DataFrame (diseaseLabels,columns=['diseaseLabels'])
diseasesNames.head()
# df = DataFrame (your_list,columns=['Column_Name'])

# [i. for i in diseaseLabels.columns]
# diseasesNames.head()
```

```python
x, y = diseases, diseaseLabels
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(len(x_train.iloc[0]),)),
    Dense(32, activation='relu'),
    Dense(1)]
)
```

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(),
              loss=tf.keras.losses.MeanSquaredError(),
              metrics=['mae', 'mse'])
```

```python
model.fit(x_train, y_train, epochs=10, validation_split=0.1)
```
